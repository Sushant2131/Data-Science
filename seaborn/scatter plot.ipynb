```python
import seaborn as sns
import pandas as pd
```

```python
df=pd.read_excel('CARS.xlsx').dropna()
a=df[20:30]
a
```

```python
sns.set_style('dark')
sns.scatterplot(x='no_cylinder',y='fuel_tank_capacity',data=a
                ,hue='fuel_type',
               )
```

```python
sns.set_style('dark')
sns.scatterplot(x='no_cylinder',y='fuel_tank_capacity',s=100,data=a,hue='fuel_type',alpha=0.5,
                edgecolor='black'
               
               )
```
