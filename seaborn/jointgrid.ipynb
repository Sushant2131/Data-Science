```python
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
df=sns.load_dataset('mpg').dropna()
a=df[:10]
a
```

```python
g=sns.JointGrid(data=df,x='weight',y='mpg')
g.plot(sns.kdeplot,sns.histplot)
```

```python
g=sns.JointGrid(data=df,x='weight',y='mpg')
g.plot_joint(sns.scatterplot)
g.plot_marginals(sns.kdeplot)
```

```python
g=sns.JointGrid()
x,y=df.weight,df.mpg
sns.regplot(x=x,y=y,ax=g.ax_joint)
sns.histplot(x=x,ax=g.ax_marg_x )
sns.kdeplot(ax=g.ax_marg_y,y=y,fill=True)
```
