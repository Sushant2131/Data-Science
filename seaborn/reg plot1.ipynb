```python
import seaborn as sns
from matplotlib import pyplot as plt
import pandas as pd
```

```python
df =pd.read_excel("CARS.xlsx").dropna()
a=df[:50]
a
```

```python
sns.regplot(data=a,x='no_cylinder',y='fuel_tank_capacity')
```

```python
sns.regplot(data=a,x='no_cylinder',y='fuel_tank_capacity',fit_reg=False)
```

```python

sns.regplot(data=a,x='no_cylinder',y='fuel_tank_capacity',scatter=False)
```

```python
import numpy as np
sns.regplot(data=a,x='no_cylinder',y='fuel_tank_capacity',x_estimator=np.median)
```

```python
sns.regplot(data=a,x='no_cylinder',y='fuel_tank_capacity',robust=True)
```

```python
sns.regplot(data=a,x='no_cylinder',y='fuel_tank_capacity',order=2)
```

```python
sns.regplot(data=a,x='no_cylinder',y='fuel_tank_capacity',logistic=False)
```

```python
sns.regplot(data=a,x='no_cylinder',y='fuel_tank_capacity',lowess=True)
```

```python
sns.regplot(data=a,x='no_cylinder',y='fuel_tank_capacity',scatter_kws={'alpha':0.5,'color':'red'})
```
