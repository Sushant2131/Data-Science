```python
import seaborn as sns
from matplotlib import pyplot as plt
import numpy as np
```

```python
df=sns.load_dataset('tips').dropna()
df.head()
```

```python
sns.pairplot(df)
```

```python
sns.pairplot(df,diag_kind='hist')
```

```python
sns.pairplot(df,kind='reg')
```

```python
sns.pairplot(df,kind='reg',hue='day')
```

```python
sns.pairplot(df,hue='day',y_vars=['tip'],x_vars=['size','total_bill'])
```

```python
from ipywidgets import interactive
def size(height=2.5,aspect=1):
    sns.pairplot(df,height=height,aspect=aspect)
    
```

```python
interactive(size,height=(1,3.5,0.5),aspect=(0.5,2,0.25))
```
