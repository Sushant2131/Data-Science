```python
import seaborn as sns
from matplotlib import pyplot as plt
import pandas as pd
import numpy as np
```

```python
penguins = sns.load_dataset('penguins').dropna()
a=penguins[:300]
```

```python
penguins.shape
```

```python
df = a.columns[2:6]

df

sns.clustermap(a[df],annot=True,fmt='.0f',standard_scale=1);
```

```python
sns.clustermap(a[df],annot=True,fmt='.0f',standard_scale=1,method='single');
```

```python
species_colors=penguins.species.map({
    'Adelie': 'blue',
    'Chinstrap': 'red',
    'Gentoo': 'green'
})

sns.clustermap(a[df], standard_scale=1, row_colors=species_colors);
```
