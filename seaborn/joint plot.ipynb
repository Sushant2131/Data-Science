```python
import seaborn as sns
from matplotlib import pyplot as plt
import matplotlib
```

```python
df=sns.load_dataset('geyser')
df.head()
```

```python
sns.jointplot(x='waiting',y='kind',data=df)
```

```python
sns.jointplot(x='waiting',y='duration',data=df,kind='kde')
```

```python
sns.jointplot(x='waiting',y='duration',data=df,kind='hex')
```

```python
sns.jointplot(x='waiting',y='duration',data=df,kind='resid')
```

```python
sns.jointplot(x='waiting',y='duration',data=df,kind='reg',joint_kws={'color':'red','marker':5,
                                                                     'ci':None}
              ,marginal_kws={'color':'green'})
```
