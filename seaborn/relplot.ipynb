```python
import seaborn as sns
from matplotlib import pyplot as plt 
import numpy as np
```

```python
df=sns.load_dataset('titanic').dropna()
df.head()
```

```python
sns.relplot(x='age',y='fare',data=df,hue='class',style='class')
```

```python
data=df.groupby('class').count()
data
```

```python
df.groupby('alive').count()
```

```python
sns.relplot(x='age',y='fare',data=df,hue='class',style='class',kind='line')
```

```python
df1=sns.relplot(x='age',y='fare',data=df,row='class',col='sex',hue='alive',style='alive')
df1.set_titles(row_template='{row_name} class',col_template='{col_name}')

df1.add_legend()
```
