```python
import seaborn as sns
import pandas as pd
```

```python
df=pd.read_excel('CARS.xlsx').dropna()
a=df
a
```

```python
data=a
sns.swarmplot(x='fuel_type',y='max_power_bhp',data=a,hue='no_cylinder',dodge=True)
```

```python
sns.boxplot(x='fuel_type',y='no_cylinder',data=a)
sns.swarmplot(x='fuel_type',y='no_cylinder',data=a,color='black',alpha=0.5)
```

```python
sns.violinplot(x='fuel_type',y='no_cylinder',data=a)
sns.swarmplot(x='fuel_type',y='no_cylinder',data=a,color='black',alpha=0.5,edgecolor='black')
```
