```python
import seaborn as sns
from matplotlib import pyplot as plt
import matplotlib
```

```python
df=sns.load_dataset('penguins')
a=df[:30]
a
```

```python
df1=sns.FacetGrid(df,row='species')
df1.map(sns.regplot,x='flipper_length_mm',y='body_mass_g',data=a)
```

```python
df1=sns.FacetGrid(df,col='species',row='island',sharey=False,hue='sex')
df1.map_dataframe(sns.scatterplot,x='flipper_length_mm',y='body_mass_g',data=a)
df1.set_axis_labels('birds','weight')
df1.set_titles(col_template='{col_name} breed',row_template='{row_name} island')
df1.add_legend()
```
