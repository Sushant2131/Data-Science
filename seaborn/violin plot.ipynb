```python
import seaborn as sns
import pandas as pd
```

```python
df=pd.read_excel('CARS.xlsx').dropna()
df.head()
```

```python
data=df
sns.violinplot(x='fuel_type',y='seating_capacity',data=df,hue='transmission_type')
```

```python
sns.violinplot(x='fuel_type',y='seating_capacity',
               data=df[df.transmission_type.isin(['Automatic','Manual'])],
               hue='transmission_type')
```

```python
sns.violinplot(x='fuel_type',y='seating_capacity',
               data=df[df.transmission_type.isin(['Automatic','Manual'])],
               hue='transmission_type',
              split=True)
```

```python
sns.violinplot(x='fuel_type',y='seating_capacity',
               data=df[df.transmission_type.isin(['Automatic','Manual'])],
               hue='transmission_type',
              split=True,
              inner='quartile')
```
