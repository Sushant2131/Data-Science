```python
import seaborn as sns
from matplotlib import pyplot as plt
import pandas as pd
import numpy as np
```

```python
df=sns.load_dataset('mpg')
df.head()
df.groupby('origin').cylinders.value_counts()
```

```python
A=(df.groupby('origin').cylinders.value_counts().unstack().fillna(0))
```

```python
sns.heatmap(A)
```

```python
df.corr()
```

```python
sns.heatmap(A.corr())
```

```python
sns.heatmap(A.corr(),cmap='Blues')
```

```python
sns.heatmap(A.corr(),cmap='RdBu',center=0)
```

```python
sns.heatmap(A.corr(),cmap='RdBu',center=0,annot=True,linewidth=1,linecolor='black',square=True)
```
