```python
import pandas as pd
india=pd.DataFrame({
    'city':['mumbai','delhi','banglore'],
    'temperture':[33,29,26],
    'wheather':['sunny','sunny','cloudy']
})
india
```

```python
us=pd.DataFrame({
    'city':['newyork','chicago','orland'],
    'temperture':[20,21,22],
    'wheather':['sunny','sunny','rainy']
})
us
```

```python
df=pd.concat([india,us],ignore_index=True)
df
```

```python
df=pd.concat([india,us],keys=['india','us'])
df
```
