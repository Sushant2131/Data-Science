```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
days=[1,2,3,4,5,6,7,8]
max_t=[39,38,37,42,41,42,43,40]
min_t=[20,22,23,21,19,25,20,22]
avg_t=[25,28,28,26,20,22,11,22]
```

```python
plt.plot(days,max_t,min_t,avg_t)
```

```python
plt.xlabel('days')
plt.ylabel('temperture')
plt.title('wheather report')
plt.plot(days,max_t,label='max')
plt.plot(days,min_t,label='min')
plt.plot(days,avg_t,label='avg')
plt.legend(shadow=True,fontsize='large')
plt.grid()
```
