```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
plt.xlabel('sugar')
plt.ylabel('No. of pepoles')
plt.title('Blood sugar analysis')
blood_sugar = [113, 85, 90, 150, 149, 88, 93, 115, 135, 80, 77, 82, 129]
blood_sugar_women = [67, 98, 89, 120, 133, 150, 84, 69, 89, 79, 120, 112, 100] 
plt.hist([blood_sugar,blood_sugar_women], rwidth=0.8,bins=[80,100,125,150],label=['men','women'])
plt.legend()
plt.grid()
```
